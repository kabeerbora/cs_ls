# Creating \& Plotting rop

```python
df['rop'] = (df['Y'] - df['WS'])/df['K']
df['rop'].describe()
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(15, 6))
ax = plt.gca()
sns.lineplot(x='year', y='rop', data=df, ax=ax)
plt.title('Rate of Profit (Organized Manufacturing)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Rate of Profit', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axvspan(2001, 2012, facecolor='lightgray', alpha=0.4)
plt.text(2007.7, df['rop'].max() * 0.7, 'Increasing Contractual Share',
         fontsize=12, ha='center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.grid(True, linestyle='dashdot', alpha=1)
plt.show()
```
